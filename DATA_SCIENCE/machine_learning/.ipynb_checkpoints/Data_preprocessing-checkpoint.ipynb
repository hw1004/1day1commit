### 가공데이터 (전처리)

```python
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
import os
import pandas as pd
import numpy as np
```

```python
# 아파트 관련 데이터
df1 = pd.read_csv("C:/Users/USER/classnote/머신러닝전처리/데이터/2019년_서울_아파트매매_실거래가.csv", encoding = "cp949")
df2 = pd.read_csv("C:/Users/USER/classnote/머신러닝전처리/데이터/2019년_서울시_아파트주소.csv", encoding = "cp949")
```

```python
df1.head()
df2.head()
```

```python
df1.columns
df2.columns
```

```python
# 데이터 병합
df = pd.merge(df1, df2, left_on=['법정동', '도로명', '아파트'], right_on=['읍면동명', '도로명', '건물명'])
```

```python
df3 = pd.read_excel("C:/Users/USER/classnote/머신러닝전처리/데이터/지하철역_위경도.xlsx")
```

#### 아파트별로 가까운 지하철역을 특정한다.

```python
# 아파트명과 역명을 index로 설정
df.set_index(['아파트'], inplace=True)
df3.set_index(['역명'], inplace=True)
```

```python
df.head(1)
df3.head(1)
```

```python
# 아파트에 대한 데이터에서 위도와 경도 컬럼을 저장해줍니다.
df_location = df[['위도', '경도']]
df3_location = df3[['위도', '경도']]
```

```python
!pip install haversine
```

```python
# 거리행렬 생성
from scipy.spatial.distance import cdist
from haversine import haversine
```

```python
# df_location과 df3_location을 이용해서 아파트와 지하철역 사이의 곡선 거리 구하기
dist_mat = cdist(df_location, df3_location, metric=haversine)
```

```python
pd.DataFrame(dist_mat)
```

```python
pd.DataFrame(dist_mat).loc[682].min()  # 특정 행에서 가장 작은 값 출력
```

```python
df_test = pd.DataFrame(dist_mat, columns=df3.index, index=df.index)
df_test.head()
```
